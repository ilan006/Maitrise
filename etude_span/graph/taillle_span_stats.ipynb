```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style="darkgrid")
```

```python
csvfile = pd.read_csv('taille_span_dev.csv')
csvfile.head()
sns.set(font_scale=1)
g = sns.factorplot(x='taille', y='compte relatif',  data=csvfile, palette='plasma_r', kind='bar', aspect = 2.2)
g.set_xticklabels(step=3)
plt.title("Distribution des tailles de span (Dev)" , size = 15)
g.set_xlabels(text="taille (en mot)")
```

```python
csvfile = pd.read_csv('taille_span_train.csv')
csvfile.head()
sns.set(font_scale=1)
g = sns.factorplot(x='taille', y='compte relatif',  data=csvfile, palette='plasma_r', kind='bar', aspect = 2.2)
g.set_xticklabels(step=3)
plt.title("Distribution des tailles de span (Train)" , size = 15)
g.set_xlabels(text="taille (en mot)")
# g.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
```

```python
csvfile = pd.read_csv('taille_span_dev_en_fonction_des_questions.csv')
csvfile.head()
sns.set(font_scale=3)
g = sns.factorplot('taille','compte relatif',hue="type question", palette='tab20',kind='point', data=csvfile,size=15, aspect=2.5)
plt.title("Taille du span en fonction du type de question" , size = 40)
g.set_xticklabels(step=3)
g.set_xlabels(text="taille (en mot)")
```

```python
csvfile = pd.read_csv('position_span_words_dev.csv')
csvfile.head()
sns.set(font_scale=3)
g = sns.factorplot('position relative','comptage relatif',hue="class", palette='tab20',kind='point', data=csvfile,size=15, aspect=2.5)
plt.title("position relative du span par rapport aux mots(dev)" , size = 40)
g.set_xticklabels(step=1)
g.set_xlabels(text="position relative (en mot)")
```
